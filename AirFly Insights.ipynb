## Data Cleaning & Preprocessing

```python
# Delay columns
delay_cols = [
    'DEPARTURE_DELAY', 'ARRIVAL_DELAY', 'AIR_SYSTEM_DELAY', 'SECURITY_DELAY',
    'AIRLINE_DELAY', 'LATE_AIRCRAFT_DELAY', 'WEATHER_DELAY'
]

for col in delay_cols:
    df[col] = df[col].fillna(0)
```

```python
# Cacellation_ reason column
df.loc[df['CANCELLED'] == 0, 'CANCELLATION_REASON'] = 'Not cancelled'
```

```python
# Tail_Number column
df['TAIL_NUMBER']=df['TAIL_NUMBER'].fillna('unkown')
```

```python
# Remove duplicated cols
df.drop_duplicates(inplace=True)
```

```python
# create Date column
cols = ['YEAR','MONTH','DAY']

for col in cols:
    df[col] = df[col].astype(str).str.strip().astype(int)
    
df['DATE']=pd.to_datetime(df[['YEAR','MONTH','DAY']])
```

```python
df.DATE.dtype
```

```python
df[['YEAR','MONTH','DAY']].dtypes
```

```python
# fixed the value in date time 
time_cols = [
    'SCHEDULED_DEPARTURE',
    'DEPARTURE_TIME',
    'SCHEDULED_ARRIVAL',
    'ARRIVAL_TIME'
]

for col in time_cols:
    
    df[col] = pd.to_numeric(df[col], errors='coerce')
    
    mask_2400 = df[col] == 2400
    
    # create temporary date
    temp_date = df['DATE'].copy()
    temp_date.loc[mask_2400] += pd.Timedelta(days=1)
    
    df.loc[mask_2400, col] = 0
    
    df[col] = df[col].astype('Int64').astype(str).str.zfill(4)
    
    df[col] = pd.to_datetime(
    temp_date.astype(str) + ' ' + df[col],
    format='%Y-%m-%d %H%M',
    errors='coerce'
    )
    print('done')
```

```python
df.head()
```

```python
month_map={1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'July',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}

df['MONTH'] = df['MONTH'].map(month_map)
```

```python
airline_map = {
    'AA': 'American Airlines',
    'AS': 'Alaska Airlines',
    'B6': 'JetBlue Airways',
    'DL': 'Delta Air Lines',
    'EV': 'ExpressJet Airlines',
    'F9': 'Frontier Airlines',
    'HA': 'Hawaiian Airlines',
    'MQ': 'American Eagle Airlines',
    'NK': 'Spirit Airlines',
    'OO': 'SkyWest Airlines',
    'UA': 'United Airlines',
    'US': 'US Airways',
    'VX': 'Virgin America',
    'WN': 'Southwest Airlines'
}

df['AIRLINE'] = df['AIRLINE'].map(airline_map)
```

```python
cancelation_map = {
    'A': 'Carrier',
    'B': 'Weather',
    'C': 'National Air System',
    'D': 'Security',
    'Not cancelled': 'Not Cancelled'
}

df['CANCELLATION_REASON'] = df['CANCELLATION_REASON'].map(cancelation_map)
```

```python
df['ROUTE'] = df['ORIGIN_AIRPORT'].astype(str) + " → " + df['DESTINATION_AIRPORT'].astype(str)
```

```python
df['FLIGHT_STATUS']='normal'
df.loc[df['CANCELLED']==1,'FLIGHT_STATUS']='cancelled'
df.loc[df['DIVERTED']==1,'FLIGHT_STATUS']='diverted'
```

```python
df['DEP_HOUR'] = df['DEPARTURE_TIME'].dt.hour
df['ARR_HOUR'] = df['ARRIVAL_TIME'].dt.hour
```

```python
df['PART_OF_DAY'] = pd.cut(
    df['DEP_HOUR'],
    bins=[0,6,12,18,24],
    labels=['Night','Morning','Afternoon','Evening'],
    right=False
)
```

```python
# create Arrival status col
df['ARR_STATUS'] = np.select(
    [
        df['CANCELLED'] == 1,
        df['DIVERTED'] == 1,
        df['ARRIVAL_DELAY'] < 0,
        df['ARRIVAL_DELAY'].between(0, 14),
        df['ARRIVAL_DELAY'] >= 15
    ],
    [
        'Cancelled',
        'Diverted',
        'Early',
        'On-Time',
        'Delayed'
    ],
    default='On-Time'
)
```

```python
df.info(memory_usage='deep')
```

```python
cat_cols = ['AIRLINE','ORIGIN_AIRPORT','DESTINATION_AIRPORT','CANCELLATION_REASON',
           'MONTH','DAY','DAY_OF_WEEK','ROUTE','FLIGHT_STATUS','TAIL_NUMBER']

for col in cat_cols:
    df[col] = df[col].astype('category')
```

```python
float_cols = ['TAXI_OUT','SCHEDULED_TIME','TAXI_IN',
    'AIR_SYSTEM_DELAY','SECURITY_DELAY',
    'AIRLINE_DELAY','LATE_AIRCRAFT_DELAY',
    'WEATHER_DELAY'
]

for col in float_cols:
    df[col] = df[col].astype('float32')

df['CANCELLED'] = df['CANCELLED'].astype('int8')
df['DIVERTED'] = df['DIVERTED'].astype('int8')

df['FLIGHT_NUMBER'] = df['FLIGHT_NUMBER'].astype('int32')
df['DISTANCE'] = df['DISTANCE'].astype('int32')

df['DEP_HOUR'] = df['DEP_HOUR'].astype('Int8')
df['ARR_HOUR'] = df['ARR_HOUR'].astype('Int8')

df['ARR_STATUS'] = df['ARR_STATUS'].astype('category')
```

#### Top Origin Destination Pairs

```python
top_origin_destination_pairs=(
    df.groupby(by=['ORIGIN_AIRPORT','DESTINATION_AIRPORT'],observed=True)
    .size()
    .sort_values(ascending=False)
    .reset_index(name='FLIGHT_COUNT')
    .head(10)
)
top_origin_destination_pairs
```

```python
# Create pivot table
route_delay = df.pivot_table(
    values='ARRIVAL_DELAY',
    index='ORIGIN_AIRPORT',
    columns='DESTINATION_AIRPORT',
    aggfunc='mean'
)

plt.figure(figsize=(14,10))
sns.heatmap(route_delay, cmap='coolwarm', center=0)

plt.title('Average Arrival Delay Heatmap (Origin vs Destination)')
plt.xlabel('Destination Airport')
plt.ylabel('Origin Airport')
plt.show()
```

```python
busiest_airports=(
    df.groupby(by='ORIGIN_AIRPORT',observed=True)
    .size()
    .reset_index(name='FLIGHT_COUNT')
    .sort_values(by='FLIGHT_COUNT',ascending=False)
    .head(10)
)
busiest_airports
```
